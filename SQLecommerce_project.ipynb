```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the DataFrames 'order_data' and 'product_data' loaded
# order_data = pd.read_csv('path_to_your_order_data.csv')  # Uncomment if loading from CSV
# product_data = pd.read_csv('path_to_your_product_data.csv')  # Uncomment if loading from CSV

# 1. Merge order_data with product_data to get original and selling prices
merged_data = order_data.merge(product_data[['product_id', 'original_price', 'selling_price']], on='product_id')

# 2. Calculate profit per unit and total profit for each product
merged_data['profit_per_unit'] = merged_data['selling_price'] - merged_data['original_price']
merged_data['total_profit'] = merged_data['profit_per_unit'] * merged_data['quantity']

# 3. Aggregate total profit for each product
product_profit = merged_data.groupby('product_id').agg({'total_profit': 'sum'}).reset_index()

# 4. Merge with product_data to get product names
product_profit = pd.merge(product_profit, product_data[['product_id', 'product_name']], on='product_id')

# 5. Sort by total profit and get top products (e.g., top 10)
top_products_by_profit = product_profit.sort_values(by='total_profit', ascending=False).head(10)

# 6. Plotting the top products by total profit
plt.figure(figsize=(12, 6))
plt.bar(top_products_by_profit['product_name'], top_products_by_profit['total_profit'], color='orange')
plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.title('Top Products by Total Profit')
plt.xticks(rotation=45, ha='right')  # Rotate product names for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
